# Import Libreries

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
```

# read the data 

```python
data = pd.read_csv('DATA/movies.csv')
data.head(5)
```

# Let's see if there is any missing data

```python
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
Newdf = data.dropna(axis=0)  
```

### Data types for our colums

```python
Newdf.dtypes
```

```python
# change data type of columns
Newdf['budget'] = Newdf['budget'].astype('int64')

Newdf['gross'] = Newdf['gross'].astype('int64')
```

```python
Newdf.head()
```

```python
# Create correct Year column
Newdf['yearcorrect'] = Newdf['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
Newdf.sort_values(by=['gross'], inplace=False, ascending=False)
```

### Drop any duplicates

```python
Newdf.drop_duplicates()
```
